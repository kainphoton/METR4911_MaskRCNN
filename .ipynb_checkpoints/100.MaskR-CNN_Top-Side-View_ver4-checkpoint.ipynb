# 0. Import Modules and setup parameters

```python
import cv2, os
import numpy as np
import math as m
from matplotlib import pyplot as plt
%matplotlib inline
print(os.getcwd())
```

```python
########################### Fruit/Vegetable/Object Parameters ###########################
# coco_labels.txt, Labels for fruit/vegetable (6 labels) 
LABELS = {0: "Person", 51: "Banana", 52: "Apple", 54: "Orange", 55: "Broccoli", 56: "Carrot"}

# Color in BGR in CV2 not RGB, match color with object's color
COLORS = {"Person": (179,179,255), "Banana": (72,216,255), "Apple": (0,8,255), 
         "Carrot": (33,145,237), "Orange": (0,126,255), "Broccoli": (0,255,74)}

# Density (g/cm3)
APPLE_DENSITY = 0.53
BANANA_DENSITY = 0.63
ORANGE_DENSITY = 0.76
# CARROT_DENSITY = 1   # TODO: need to find out
# BROCCOLI_DENSITY = 1 # TODO: need to find out

# Calories per gram
APPLE_CONVERT_TO_CALORIE = 52 / 100        # 52 calorie per 100 g
BANANA_CONVERT_TO_CALORIE = 89 / 100       # 89 calorie per 100 g
ORANGE_CONVERT_TO_CALORIE = 47 / 100       # 47 calorie per 100 g
CARROT_CONVERT_TO_CALORIE = 41 / 100       # 41 calorie per 100 g
BROCCOLI_CONVERT_TO_CALORIE = 34 / 100     # 34 calorie per 100 g

# Finger Size (cm)
INDEX_FINGER_WIDTH = 1.48    # 14.8mm
INDEX_FINGER_HEIGHT = 1.185  # 11.85mm

# Object Detect Mask and Score Threshold
MASK_THRESHOLD = 0.65
SCORE_THRESHOLD = 0.5

########################### Window Parameters ###########################
# Input Resize Image
IMG_WIDTH, IMG_HEIGHT  = (640, 480)

#define plot size for all plots
plt.rcParams['figure.figsize'] = [6.4*1.75, 4.8*1.75]
```

# 1. Loading Mask RCNN

```python
# Get file path of weights(.pb) and config (.pbtxt)
frozen_pb = os.path.join("cv_dnn", "mask_rcnn_inception_v2_coco_2018_01_28", "frozen_inference_graph.pb")
pb_txt = os.path.join("cv_dnn", "mask_rcnn_inception_v2_coco_2018_01_28.pbtxt")

print("Weights:", frozen_pb)
print("Config:", pb_txt)

# Load the weights and the config of Mask RCNN
net = cv2.dnn.readNetFromTensorflow(frozen_pb, pb_txt)
```

## 1.1 Object Detection and Mask Segmentation Function

```python
""" 
    Modified Source code from: https://www.youtube.com/watch?v=8m8m4oWsp8M
    YouTube title: Instance Segmentation MASK R-CNN | with Python and Opencv
    YouTube Name: Pysource
    Date: May 18, 2021
"""
def object_detection_mask(img):
    """
    Passes the img into the MaskRCNN model to get the object detection boxes and its masks.
    
        Parameters:
                    img (numpy.ndarray): Original image

        Returns:
                    img (numpy.ndarray): Image with object detection boxes+labels drawn on top
                    black_image (numpy.ndarray): Image segmentation mask of each detected object
                    box_list (list: numpy.ndarray): List of detected boxes
    """
    height, width, _ = img.shape  # (row, column)
    box_list = []
    
    ########################### Generate random colors and create black image ###########################
    black_image = np.zeros((height, width, 3), np.uint8)
    
    ########################### Detect objects ###########################
    blob = cv2.dnn.blobFromImage(img, swapRB=True)  # Get blob of input image
    net.setInput(blob)
    
    # boxes.shape (1,1,100,7),  mask.shape (100,90,15,15)
    boxes, masks = net.forward(["detection_out_final", "detection_masks"])  # Actual prediction (takes time)
    detection_count = boxes.shape[2]  # number of objects detected, 100 is the limit
    
    ############################ For each objects detected, draw box and mask ###########################
    for i in range(detection_count):
        # box [0,label,confidence,x1,y1,x2,y2] coods are normalised to 0-1 so multiply 
        #     the coods with image's width and height
        box = boxes[0, 0, i]
        class_id = int(box[1])  # Label/Class number, refer to coco_labels.txt
        score = box[2]          # Confidence percentage
        
        # If score below (is part of label) or not part of labels want to detect skip
        if (class_id not in LABELS) or (score < SCORE_THRESHOLD):
            continue

        color = COLORS[LABELS[class_id]] # Color of object, defined in section 0
                
        ############################ Get Box Coordinates and Draw Box and Label ############################
        x = int(box[3] * width)    # Top left
        y = int(box[4] * height)   # Top left
        x2 = int(box[5] * width)   # Bottom right
        y2 = int(box[6] * height)  # Bottom right
        
        new_box = list(box.copy())
        new_box[3:] = (x, y, x2, y2)
        new_box[1] = LABELS[int(class_id)]
        box_list.append(new_box) # Get the box data
        
        # Draw rectangle box around detected object
        line_thickness = 2
        pad_x, pad_y = 2, 4
        cv2.rectangle(img, (x,y), (x2,y2), color, line_thickness) 
        
        # Draw Label's rectangle box and Add Text
        label_box_height = 20   # Change Height of Label Box
        label_box_width = 75    # Change Width of Label Box
        
        # Positions for box
        label_x = x - line_thickness // 2 
        label_y = y - label_box_height
        label_x2 = x + label_box_width
        label_y2 = y
        
        # Draw Box
        cv2.rectangle(img, (label_x,label_y), (label_x2,label_y2), color, cv2.FILLED)
        
        # Draw Text
        font = cv2.FONT_HERSHEY_SIMPLEX
        cv2.putText(img, LABELS[class_id], (x + pad_x, y - pad_y), font, 0.5, (0,0,0), 1, cv2.LINE_AA)
        
        ############################ Get the mask ############################
        roi = black_image[y: y2, x: x2]
        
        roi_height, roi_width, _ = roi.shape
        mask = masks[i, int(class_id)]  # mask.shape -> (15, 15)
        mask = cv2.resize(mask, (roi_width, roi_height))    # Resize from (15,15) to actual size
        
        # 0.5 is threshold of confidence score of each pixel? so over 0.5 becomes 255 and 0 otherwise
        _, mask = cv2.threshold(mask, MASK_THRESHOLD, 255, cv2.THRESH_BINARY)
        
        ############################ Get Mask Coordinates ############################
        # Extract boundary of object
        contours, _ = cv2.findContours(np.array(mask, np.uint8), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)         
        
        ############################ Draw and Fill Polygons ############################
        for cnt in contours:
            cv2.fillPoly(roi, [cnt], color)
    
    ############################ Display output image on computer ############################
    cv2.imshow("Frame", img)
    cv2.imshow("Black Image", black_image)
    
    return (img.copy(), black_image.copy(), box_list)
```

# 2. Detect Static Image

```python
# Load image
# img_file = "banana.d8fbcaf6-03f4-11ed-a1f3-95c0acd580f9.jpg"
# img_path = os.path.join("old_test_train_images_xml", "Fruit_Vegetable_15-July-22", "collectedimages", "banana", img_file)

img_file = "orange-finger-original.jpg"
img_path = os.path.join("WebCam_images", img_file)

print(img_path)
```

```python
# Read image and resize image to (640, 480) or other dimensions
img = cv2.imread(img_path)
img = cv2.resize(img, (IMG_WIDTH,IMG_HEIGHT), interpolation = cv2.INTER_AREA)

original_img = img.copy() # Vanilla pure image, no overlays, boxes, labels, etc
overlay = img.copy() # For mask overlay
height, width, _ = img.shape

# 0 = open until any key is pressed, then close, any other number, in milliseconds, opens window in that duration
cv2.imshow("Image ({},{})".format(width, height), img)
cv2.waitKey(0)  
cv2.destroyAllWindows()

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # Inline show image in notebook

print("Width:", width)
print("Height:", height)
```

```python
# For Static Image Only
object_detect_img, segmentation_img, box_list = object_detection_mask(img)
cv2.waitKey(0)
cv2.destroyAllWindows() 

# Object Detection Box Image
plt.imshow(cv2.cvtColor(object_detect_img, cv2.COLOR_BGR2RGB))
```

```python
# Image Segementation Image
plt.imshow(cv2.cvtColor(segmentation_img, cv2.COLOR_BGR2RGB))
```

```python
# Overlay mask into object detection image
alpha = 0.4  # Transparency factor

# Third parameter is image object detect box
mask_overlay = cv2.addWeighted(segmentation_img, alpha, object_detect_img, 1 - alpha, 0)
  
cv2.imshow("Mask Overlay", mask_overlay)
cv2.waitKey(0)
  
cv2.destroyAllWindows()
plt.imshow(cv2.cvtColor(mask_overlay, cv2.COLOR_BGR2RGB))
```

# 3. Real Time Video Object Detection and Segmentation

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame = cap.read()
    
    # object_detection_mask function returns -> (object_detect_img, segmentation_img)
    object_detect_img, segmentation_img = object_detection_mask(frame)
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```

# 4. Get "area" of each object detected

```python
# Available variables: box_list, segmentation_img, mask_overlay
# box [0,label,confidence,x1,y1,x2,y2] x1,y1,x2,y2 have been updated from function
window_x, window_y = 0, 0

for i in range(len(box_list)):
    # Display window of the detected object only
    box = box_list[i]
    x1,y1,x2,y2 = box[3:]
    object_img = img[y1:y2, x1:x2].copy()
    window_name_od = "Object Detect: {}".format(box[1])
    cv2.imshow(window_name_od, object_img)
    cv2.moveWindow(window_name_od, window_x, window_y)
    
    ### Display window of the image segmentaiton of detected object only
    segment_img = segmentation_img[y1:y2, x1:x2].copy()
    window_name_segment = "Image Segmentation: {}".format(box[1])
    cv2.imshow(window_name_segment, segment_img)
    cv2.moveWindow(window_name_segment, window_x + (x2-x1), window_y)
    
    # Display window of the mask overlay
    mask_overlay_object = mask_overlay[y1:y2, x1:x2].copy()
    window_name_overlay = "Mask Overlay: {}".format(box[1])
    cv2.imshow(window_name_overlay, mask_overlay_object)
    cv2.moveWindow(window_name_overlay, window_x + (x2-x1)*2, window_y)
      
    cv2.waitKey(0)  
    cv2.destroyAllWindows()
```

```python
# Convert segment_img to greyscale to get single color channel, as cv2.countNonZero() only works with single channel
gray = cv2.cvtColor(segment_img, cv2.COLOR_BGR2GRAY)
cv2.imshow("Gray", gray)
cv2.waitKey(0)  
cv2.destroyAllWindows()

num_pixels = cv2.countNonZero(gray)
print(f'Number of pixels: {num_pixels}')
plt.imshow(cv2.cvtColor(gray, cv2.COLOR_BGR2RGB))
```

```python
# Assumes only finger up to wrist (right to left)
# Cropped finger image segmentation, box_list, assumes only one person object detected
# box [0,label,confidence,x1,y1,x2,y2] eg. [0.0, 'Person', 0.95590365, 1, 199, 249, 331]
box_person = []
for box in box_list:
    if box[1] == 'Person':
        box_person.extend(box)
        break
        
person_x1, person_y1, person_x2, person_y2 = box_person[3:]
box_person_img_mask = segmentation_img[person_y1:person_y2, person_x1:person_x2].copy()
window_name_segment = f"Image Segmentation: {box_person_img_mask}"

box_person_img_mask = cv2.cvtColor(box_person_img_mask, cv2.COLOR_BGR2GRAY)
box_person_img_mask_no_lines = box_person_img_mask.copy() # image with no lines
person_box_height, person_box_width = box_person_img_mask.shape

# Draw vertical line
xline_ratio = 0.1 # offset line ratio to the left
height_ratio_threshold = 0.05 # ratio of window height threshold for line1
vertical_line_pixel_threshold = round(person_box_height * height_ratio_threshold)

### Scan vertical line from right to left, stop when the vertical line contains some pixel colors
for x in range(person_box_width, -1, -1):
    single_vertical_line_pixel = np.count_nonzero(box_person_img_mask[:, x:x+1])
    if single_vertical_line_pixel > vertical_line_pixel_threshold:
        x_start = x
        break

# First line touch edge of finger
line1_start = (x_start, 0)
line1_end = (x_start, person_box_height)
cv2.line(box_person_img_mask, line1_start, line1_end, (255,255,255), 1)  # Draw outer line

# Second line offset to the left by window ratio to get location of finger height
x_offset = round(person_box_width * xline_ratio)
line2_start = (line1_start[0] - x_offset, 0)
line2_end = (line1_start[0] - x_offset, person_box_height)
cv2.line(box_person_img_mask, line2_start, line2_end, (255,255,255), 1)  # Draw inner line

# Get number of pixels of finger in inner line
inner_single_vertical_line_pixel = np.count_nonzero(box_person_img_mask_no_lines[:, line2_start[0]:line2_start[0]+1])

# Show window
cv2.imshow(window_name_segment, box_person_img_mask)
cv2.waitKey(0)  
cv2.destroyAllWindows()
plt.imshow(cv2.cvtColor(box_person_img_mask, cv2.COLOR_BGR2RGB))

print("Window-Width:", person_box_width)
print("Window-Height:", person_box_height)
print("Vertical Line Width Point:", x_start)
print("Offset Width Vertical Line:", x_offset)
print("Vertical Lines Width Ratio Offset:", xline_ratio, '\n')
print("Number of pixels in finger's height from inner vertical line:", inner_single_vertical_line_pixel) # Fix
```

```python
# inner_single_vertical_line_pixel -> 26 pixels | INDEX_FINGER_WIDTH -> 1.48cm
# Number of pixels per 1cm 
pixels_length_per_cm = round(inner_single_vertical_line_pixel / INDEX_FINGER_WIDTH)

# Number of pixels square: 1cm * 1cm = 1cm2
pixels_per_cm_squared = pixels_length_per_cm * pixels_length_per_cm
```

```python
## Show 1cm^2 rectangle overlay (***TESTING***)

# Third parameter is image object detect box
x1_square, y1_square = (0,0)
x2_square, y2_square = (pixels_length_per_cm, pixels_length_per_cm)
segment_calibration_square = segmentation_img.copy()
cv2.rectangle(segment_calibration_square, (x1_square,y1_square), (x2_square,y2_square), (0,255,0), cv2.FILLED)

mask_overlay_calibration_square = cv2.addWeighted(segment_calibration_square, alpha, object_detect_img, 1 - alpha, 0)
  
cv2.imshow("1cm^2 calibration square", mask_overlay_calibration_square)
cv2.waitKey(0)
  
cv2.destroyAllWindows()
plt.imshow(cv2.cvtColor(mask_overlay_calibration_square, cv2.COLOR_BGR2RGB))
print("Number of pixels in 1cm line:", pixels_length_per_cm)
print("Number of pixels in 1cm x 1cm square:", pixels_per_cm_squared)
```

```python
# Get cm^2 area for each object (exclusive of person) - segmentation_img - box_list
img_segment_objects_seperate_ls = []
num_of_pixels_ls = []
for i in range(len(box_list)):
    box = box_list[i]
    x1,y1,x2,y2 = box[3:]
    
    img_segment_object = segmentation_img[y1:y2, x1:x2].copy()
    gray_color_object = cv2.cvtColor(img_segment_object, cv2.COLOR_BGR2GRAY)
    
    img_segment_objects_seperate_ls.append(gray_color_object)
    
    cv2.imshow("Gray object", gray_color_object)
    cv2.waitKey(0)  
    cv2.destroyAllWindows()

    num_pixels = cv2.countNonZero(gray_color_object)
    num_of_pixels_ls.append(num_pixels)
    
    area = num_pixels / pixels_per_cm_squared
        
    print(f'Number of pixels in {box[1]}: {num_pixels}')
    print(f'Area (cm^2) of object in {box[1]}: {area}')
    print()
    plt.figure()
    plt.title(f"{box[1]} | Number of pixels: {num_pixels} | Area (cm^2): {area}")
    plt.imshow(cv2.cvtColor(gray_color_object, cv2.COLOR_BGR2RGB))
```

```python
# TESTING PURPOSES, not suitable for new images, pixels_per_cm_squared
orange_pixels = num_of_pixels_ls[1]
orange_area = orange_pixels / pixels_per_cm_squared
orange_area
```

```python
# Spherical/ellipsoid shape for ORANGE estimation
R = (orange_area / m.pi)**0.5  # Area of circle equation
V_sphere_orange = (4/3) * R * orange_area # cm^3
```

```python
# Convert volume (cm^3) -> mass (g) -> calories (cal/g)

# m = p * v
orange_mass = ORANGE_DENSITY * V_sphere_orange
print(f'Mass of orange: {round(orange_mass)} grams')

# calories = cal/g * m
orange_calory = ORANGE_CONVERT_TO_CALORIE * orange_mass
print(f'Calories of orange: {round(orange_calory)} calories')
```

# Class for each object with full information

```python
class Item:
    """
    Object detected in images with their full information.
    """
    def __init__(self, full_original_img, full_segmentation_img, box):
        # Full original image
        self.full_original_img = full_original_img
        
        # Full segmentation image
        self.full_segmentation_img = full_segmentation_img
        
        # box [0,label,confidence,x1,y1,x2,y2]
        self.box = box  
        
        # Extract box values
        # eg. box = [0.0, 'Person', 0.95590365, 1, 199, 249, 331]
        self.name = box[1]
        self.confidence = box[2]
        self.x1, self.y1 = box[3:5]
        self.x2, self.y2 = box[5:]
        
        # Initialize after getting calibration pixel per cm (finger)
        self.pixels_per_cm = None
        self.pixels_per_1cm_square = None
        
        # Update after getting calibration pixel per cm (finger)
        self.side_pixel_area = None
        self.side_area = None
        
        # Initialize to None, to be calculated
        self.mass = None
        self.calorie = None
        
    def get_name(self):
        return self.name
    
    def get_confidence(self):
        return self.confidence
    
    def get_rectangle_coods_top_left(self):
        return self.x1, self.y1
    
    def get_rectangle_coods_bot_right(self):
        return self.x2, self.y2
    
    # Need to show orignal image, not the box stuff and label stuff
    def show_original_image(self):
        cv2.imshow("Original Image", self.full_original_img)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
        
    def show_segmentation_image(self):
        cv2.imshow("Segmentation Image", self.full_segmentation_img)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
    
```
